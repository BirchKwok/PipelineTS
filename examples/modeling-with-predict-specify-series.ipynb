```python
from PipelineTS.dataset import LoadMessagesSentDataSets
```

```python
import pandas as pd
time_col = 'date'
target_col = 'ta'

init_data = LoadMessagesSentDataSets()[[time_col, target_col]]
init_data[time_col] = pd.to_datetime(init_data[time_col])
```

```python
accelerator = 'cpu'
lags = 60
n = 40
quantile = 0.9
```

```python
valid_data = init_data.iloc[-n:, :]
series = init_data.iloc[-lags * 3:-lags, :]
data = init_data.iloc[:-n, :]
data.tail(5)
```

```python
series.tail()
```

# WideGBRTModel

```python
from PipelineTS.ml_model import WideGBRTModel

wide_gbrt = WideGBRTModel(time_col=time_col, target_col=target_col, lags=lags, random_state=42, 
                          quantile=quantile)
wide_gbrt.fit(data)
```

```python
wide_gbrt.predict(n, series=series)
```
